```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 3
y = 2.5
```

```python
def plot_scatter(data, label, prediction=None):
    plt.Figure(figsize=(8, 5))
    plt.scatter(data, label, label="Training Data", c="b", s=10)
    
    if prediction is not None:
        plt.scatter(data, prediction, label="prediction", c="r", s=10)
    
    plt.title("My Data")
    plt.legend()
    
plot_scatter(x, y)
```

```python
np.random.seed(47)
W1 = np.random.rand()
b1 = np.random.rand()
W2 = np.random.rand()
b2 = np.random.rand()

print(W1, b1, W2, b2)
```

```python
def forward_propagation(W1, b1, W2, b2, data):
    h1 = W1*data + b1
    y_pred = W2*h1 + b2
    
    return h1, y_pred
```

```python
h1 , y_pred = forward_propagation(W1, b1, W2, b2, x)
plot_scatter(x, y , y_pred)
print(h1, y_pred)
```

```python
L = (y_pred - y) ** 2

dL_dy = 2 * (y_pred - y)
dy_dW2 = h1
dy_dh1 = W2
dy_db2 = 1
dh1_dW1 = x
dh1_db1 = 1


dL_dW2 = dL_dy * dy_dW2
dL_db2 = dL_dy * dy_db2
dL_dW1 = dL_dy * dy_dh1 * dh1_dW1
dL_db1 = dL_dy * dy_dh1 * dh1_db1

print(dL_dW2, dL_db2, dL_dW1, dL_db1)
```

```python
lr = 0.01
W1 = W1 - lr * dL_dW1
b1 = b1 - lr * dL_db1
W2 = W2 - lr * dL_dW2
b2 = b2 - lr * dL_db2

print(W1, b1, W2, b2)
```

```python
h1 , y_pred = forward_propagation(W1, b1, W2, b2, x)
plot_scatter(x, y , y_pred)
print(h1, y_pred)
```

```python
x = 3
y = 2.5

np.random.seed(47)
W1 = np.random.rand()
b1 = np.random.rand()
W2 = np.random.rand()
b2 = np.random.rand()

print(W1, b1, W2, b2)

apochs = 100
for apoch in range(apochs):
    # Forward
    h1 , y_pred = forward_propagation(W1, b1, W2, b2, x)

    # Loss
    L = (y_pred - y) ** 2
    
    if apoch % 10 == 0:
        print(f"Apoch = {apoch}/{apochs}, L = {L}")

    # Backpropagation
    dL_dy = 2 * (y_pred - y)
    dy_dW2 = h1
    dy_dh1 = W2
    dy_db2 = 1
    dh1_dW1 = x
    dh1_db1 = 1


    dL_dW2 = dL_dy * dy_dW2
    dL_db2 = dL_dy * dy_db2
    dL_dW1 = dL_dy * dy_dh1 * dh1_dW1
    dL_db1 = dL_dy * dy_dh1 * dh1_db1

    # Optimization
    lr = 0.01
    W1 = W1 - lr * dL_dW1
    b1 = b1 - lr * dL_db1
    W2 = W2 - lr * dL_dW2
    b2 = b2 - lr * dL_db2

h1 , y_pred = forward_propagation(W1, b1, W2, b2, x)
plot_scatter(x , y , y_pred)
```
